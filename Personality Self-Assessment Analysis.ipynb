```python
# Read raw data

import pandas as pd
data = pd.read_csv("data/personality_freeform/data_clean.csv", sep='\t')
data
```

```python
# Transform some attributes

data.ix[data['engnat'] == 2, 'engnat'] = 0
data['gender'] = 'O'
data.ix[data['gender12'] == 1, 'gender'] = 'M'
data.ix[data['gender12'] == 2, 'gender'] = 'F'
data = data.drop('gender12', 1)
data
```

```python
# Combine text items

data['text'] = data['item1'] + ' ' + data['item2'] + ' ' + data['item3']
data = data.drop('item1', 1)
data = data.drop('item2', 1)
data = data.drop('item3', 1)
data
```

```python
# Define text-derived attributes

import re

data['use_period'] = 0
data['use_comma'] = 0
data['use_apostrophe'] = 0
data['use_hyphen'] = 0
data['use_slash'] = 0
data['use_question'] = 0
data['use_parentheses'] = 0
data['use_upper'] = 0
data['use_lower'] = 0
data['word_count'] = 0
data['swear_count'] = 0

for i in data.index:
    
    s = data.ix[i, 'text']
    
    if "." in s:
        data.ix[i, 'use_period'] = 1
    
    if "," in s:
        data.ix[i, 'use_comma'] = 1
    
    if "'" in s:
        data.ix[i, 'use_apostrophe'] = 1
    
    if "-" in s:
        data.ix[i, 'use_hyphen'] = 1
    
    if "/" in s:
        data.ix[i, 'use_slash'] = 1
    
    if "?" in s:
        data.ix[i, 'use_question'] = 1
    
    if "(" in s:
        data.ix[i, 'use_parentheses'] = 1
    
    if ")" in s:
        data.ix[i, 'use_parentheses'] = 1
    
    for c in s:
        if c.islower():
            data.ix[i, 'use_lower'] = 1
            break
    
    for c in s:
        if c.isupper():
            data.ix[i, 'use_upper'] = 1
            break
            
    data.ix[i, 'word_count'] = len(s.split())
    
    for w in s.lower().split():
        if ("fuck" in w) or ("shit" in w) or ("damn" in w) or ("bitch" in w) or ("cunt" in w) or ("asswhole" in w):  #sic
            data.ix[i, 'swear_count'] += 1

            
data
```
